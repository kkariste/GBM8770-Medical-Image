````python
def repartition_classes(sous_ensemble: str) -> tuple:
    """Détermine les classes.

    Args:
        sous_ensemble (str): chaîne de caractère qui représente un sous-ensemble du dataset

    Returns:
        tuple : types de classes, nombre et proportion des images dans le dataset.
    """
    classes, nombres = np.unique(dataset[sous_ensemble]["label"], return_counts=True)
    proportions = nombres / len(dataset[sous_ensemble]["label"])

    # Histogramme de la distribution des classes
    fig, ax = plt.subplots()
    ax.bar(classes, nombres, edgecolor="black")
    ax.set_title(f"Répartition des classes dans '{sous_ensemble}'")
    ax.set_xlabel("classes")
    ax.set_ylabel("Nombre d'images")
    plt.show()
    return classes, nombres, proportions

classes_train, nombres_train, proportions_train = repartition_classes("train")
print(classes_train, nombres_train, proportions_train)

classes_validation, nombres_validation, proportions_validation = repartition_classes("validation")
print(classes_validation, nombres_validation, proportions_validation)

classes_test, nombres_test, proportions_test = repartition_classes("test")
print(classes_test, nombres_test, proportions_test)

# Vérifier la représentation des classes dans les métadonnées
print(dataset["train"].features["label"])
# Obtenir les noms des classes
label_info = dataset["train"].features["label"]
class_names = label_info.names

print(f"Les noms des classes sont : {class_names}")
print(f"Classe 0 : {class_names[0]}")  # Affiche le nom pour la classe 0
print(f"Classe 1 : {class_names[1]}")  # Affiche le nom pour la classe 1
````

## Petite note sur les fonctions fournies

### `get_model_and_optimizer`
Crée un modèle et son optimiseur. Cette fonction prend 2 arguments: un float `lr` (le taux d'apprentissage ou *learning rate*) et un booléen optionnel `pretrained` (pour indiquer si le modèle a été pré-entraîné sur l'ensemble [ImageNet](https://en.wikipedia.org/wiki/ImageNet) ou s'il a été initialisé aléatoirement).

### `get_dataloaders`
Crée trois dataloaders (qui permettent d'itérer sur un dataset et qui devront être passées à `train_epoch` et `validate`) -- un pour chaque sous-ensemble (entraînement, validation, test). Cette fonction attend deux argument obligatoires: un `DatasetDict` (le dataset qu'on a créé plus haut) et un entier `batch_size` (qui détermine combien d'exemples le modèle traitera à chaque itération).

Quelques notes additionnelles sur cette fonction:
1. En général, il est préférable d'utiliser la plus grande *batch size* possible qui entre en mémoire sur votre GPU. Vous pouvez expérimenter avec différentes valeurs (typiquement par puissances de 2) jusqu'à ce que vous trouviez une valeur qui fonctionne, mais une *batch size* de 128 devrait être suffisante et entrer en mémoire de la plupart des GPU.
2. La fonction offre un paramètre optionnel `train_subset` qui est un float compris entre 0 et 1 qui vous permet de réduire la taille de l'ensemble d'entraînement.

### `train_epoch`
Entraîne le modèle sur une époque. Elle prend 3 arguments: le modèle, le dataloader d'entraînement, et l'optimiseur. Elle retourne la liste des valeurs de *loss* à chaque itération (par batch).

### `validate`
Évalue le modèle sur un ensemble de données. Elle prend 2 arguments: le modèle et un dataloader. Elle retourne deux floats et une liste d'entiers: la valeur de *loss*, la précision (*accuracy*), et les prédictions du modèle sur l'ensemble qu'on valide.

À noter que cette fonction peut techniquement être utilisée avec n'importe quel dataloader, pas nécessairement celui de validation. Cela pourrait être utile pour évaluer le modèle sur l'ensemble d'entraînement ou l'ensemble de test...

## (Sur-)entraînement d'un modèle

Une approche fréquemment recommandée pour valider qu'une architecture donnée est correctement implémentée et qu'elle est effectivement capable d'apprendre quelque chose est d'utiliser un tout petit sous-ensemble des données d'entraînement et de volontairement sur-entraîner (*overfitting*) jusqu'à convergence sur ce sous-ensemble.

**Q5**. Créez un modèle qui n'a pas été pré-entraîné (utilisez les valeurs par défaut de **pretrained** et **lr**). Sur-entraînez le sur 1% des données d'entraînement (50 époques devraient suffire). Affichez la courbe d'évolution de votre erreur (*loss*) d'entraînement. Évaluez ensuite sur le même sous-ensemble (i.e. avec le même dataloader) et affichez la *loss* et l'accuracy finales.

````python
overfit_loader, val_loader, test_loader = get_dataloaders(dataset, batch_size=128, train_subset=0.01)
# On crée un modèle non pré-entraîné
model, optimizer = get_model_and_optimizer()

# On sur-entraîne le modèle
losses = []
nombre_epoques = 50
for epoque in tqdm(range(nombre_epoques)):
    pertes_epoque = train_epoch(model, overfit_loader, optimizer)
    losses.extend(pertes_epoque)

# Courbe d'évolution de la fonction de notre erreur (loss) d'entraînement
fig, ax = plt.subplots()
ax.plot(losses)
ax.set_xlabel('Batch')
ax.set_ylabel('Loss')
ax.set_title("Évolution de la loss d'entraînement")
plt.show()

# On évalue le modèle sur le même sous-ensemble
avg_loss, accuracy, preds = validate(model, overfit_loader)
print(f"Pertes finales d'entraînement : {avg_loss}")
print(f"Accuracy finale : {accuracy}")
````

## Variation du taux d'apprentissage

<span style="color:red">**Attention !** Pour la suite du TP on veut utiliser le plus d'images possible du dataset, mais en faisant un compromis avec le temps d'entraînement. Le temps que vous écriviez votre code et que vous le testiez, vous pouvez utiliser une petite proportion des données. Un des paramètres essentiels dans l'entraînement d'un modèle est le taux d'apprentissage (*learning rate*), qui détermine la taille du pas qu'on prend lors de la descente de gradient pour optimiser un modèle.

**Q8**. Entraînez 3 modèles avec respectivement un *learning rate* de `1`, `1e-3`, et `1e-6`. Assurez-vous de laisser l'argument `pretrained` de la fonction `get_model_and_optimizer` à sa valeur par défaut de `False`. Durant l'entraînement de chaque modèle, faites une sauvegarde du modèle lorsque vous obtenez la meilleure accuracy. Affichez les courbes d'évolution de votre *loss* d'entraînement (par batch), et de votre *loss* et *accuracy* de validation (par époque).

Pour sauvegarder un modèle, vous pouvez utiliser la syntaxe suivante:
```py
torch.save(model.state_dict(), f"mon_modèle.pt")
```
La méthode `state_dict()` d'un modèle PyTorch retourne un `dict` qui contient les poids du modèle, tandis que la fonction `torch.save` permet de sauvegarder un tenseur (ou une collection de tenseurs comme un `dict`) sous un nom de fichier donné.

Pour charger un modèle à partir d'une sauvegarde, il faut d'abord créer un modèle identique, charger le `dict` des poids depuis le fichier, puis charger le `dict` de poids dans le modèle.
```py
saved_model = get_model_and_optimizer(pretrained=False)[0]
saved_state_dict = torch.load("mon_modèle.pt", map_location="cpu")
saved_model.load_state_dict(saved_state_dict)
```

Pour résoudre cet exercie, nous avons passé à chatGPT la question Q8, la syntaxe de sauvegarde et la petite note sur les fonctions fournies.

````python
train_loader, val_loader, test_loader = get_dataloaders(dataset, batch_size=128, train_subset=0.50) #0.25

# Liste des taux d'apprentissage à tester
learning_rates = [1, 1e-3, 1e-6]

# Dictionnaires pour stocker les pertes et accuracies par taux d'apprentissage
train_losses = {lr: [] for lr in learning_rates}
val_losses = {lr: [] for lr in learning_rates}
val_accuracies = {lr: [] for lr in learning_rates}

# Entraînement et validation pour chaque taux d'apprentissage
for lr in learning_rates:
    model, optimizer = get_model_and_optimizer(lr=lr)
    best_accuracy = 0
    best_model_path = f"model_lr_{lr}.pt"

    # Entraînement sur 10 époques
    for epoch in range(10):
        # Entraînement sur une époque
        epoch_train_losses = train_epoch(model, train_loader, optimizer)
        train_losses[lr].extend(epoch_train_losses)

        # Validation sur l'ensemble de validation
        val_loss, val_accuracy, _ = validate(model, val_loader)
        val_losses[lr].append(val_loss)
        val_accuracies[lr].append(val_accuracy)

        # Sauvegarde du modèle si meilleure accuracy de validation
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            torch.save(model.state_dict(), best_model_path)

        print(f"Epoch {epoch + 1} | lr={lr} | Train Loss: {epoch_train_losses[-1]:.4f} | "
              f"Val Loss: {val_loss:.4f} | Val Accuracy: {val_accuracy * 100:.2f}%")
````

````python
# Affichage des courbes d'évolution de la loss d'entraînement, loss de validation et accuracy de validation
for lr in learning_rates:
    plt.figure(figsize=(15, 5))

    # Courbe de la loss d'entraînement
    plt.subplot(1, 3, 1)
    plt.plot(train_losses[lr], label=f'lr={lr}')
    plt.xlabel('Batch')
    plt.ylabel('Training Loss')
    plt.title(f'Training Loss per Batch (lr={lr})')
    plt.legend()

    # Courbe de la loss de validation
    plt.subplot(1, 3, 2)
    plt.plot(val_losses[lr], label=f'lr={lr}')
    plt.xlabel('Epoch')
    plt.ylabel('Validation Loss')
    plt.title(f'Validation Loss per Epoch (lr={lr})')
    plt.legend()

    # Courbe de l'accuracy de validation
    plt.subplot(1, 3, 3)
    plt.plot(val_accuracies[lr], label=f'lr={lr}')
    plt.xlabel('Epoch')
    plt.ylabel('Validation Accuracy')
    plt.title(f'Validation Accuracy per Epoch (lr={lr})')
    plt.legend()

    plt.show()
````

**Q9**. Que pouvez-vous conclure sur l'impact du taux d'apprentissage sur l'entraînement de votre modèle? Quel est le meilleur taux d'apprentissage pour ce problème?


Au fur et à mesure que le taux d'apprentissage devient trop grand, le modèle s'éloigne considérablement des valeurs ou de la valeur minimale.es et oscille dans les valeurs maximales sans jamais redescendre. ## Apprentissage par transfert
Une technique standard pour améliorer les performances d'un modèle est d'utiliser l'apprentissage par transfert (*transfer learning*) où on utilise un modèle qui a préalablement été entraîné sur une grande quantité de données à résoudre une tâche connexe à la nôtre. On procède à un affinage (*finetuning*) de ce modèle sur des nouvelles données pour le spécialiser sur une nouvelle tâche.

Expliquez votre raisonnement.


Oui, nous pensons qu'utiliser un modèle pré-entraîné sur ImageNet va améliorer nos performances de classifications, mais avec certaines limites. En effet comme son nom l'indique, le transfert d'apprentissage consiste à se baser sur des expériences déjà acquises pour explorer d'autres horizons. **Q12**. Affinez pendant au moins 10 époques un modèle pré-entraîné. Créez le en utilisant l'argument `pretrained=True`. Utilisez le *learning rate* qui vous a donné les meilleures performances à l'exercice précédent. Durant l'entraînement du modèle, faites une sauvegarde du modèle lorsque vous obtenez la meilleure accuracy. Affichez les courbes d'évolution de votre *loss* d'entraînement, et de votre *loss* et *accuracy* de validation.

````python
train_loader, val_loader, test_loader = get_dataloaders(dataset, batch_size=128, train_subset=0.50) #0.25

# taux d'apprentissage
lr = 1e-3

# Dictionnaires pour stocker les pertes et accuracies par taux d'apprentissage
train_losses = []
val_losses = []
val_accuracies = []

# Entraînement et validation pour chaque taux d'apprentissage

model, optimizer = get_model_and_optimizer(lr=lr, pretrained=True)
best_accuracy = 0
best_model_path = f"model_lr_transfert.pt"

# Entraînement sur 10 époques
for epoch in range(10):
        # Entraînement sur une époque
        epoch_train_losses = train_epoch(model, train_loader, optimizer)
        train_losses.extend(epoch_train_losses)

        # Validation sur l'ensemble de validation
        val_loss, val_accuracy, _ = validate(model, val_loader)
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)

        # Sauvegarde du modèle si meilleure accuracy de validation
        if val_accuracy > best_accuracy:
            best_accuracy = val_accuracy
            torch.save(model.state_dict(), best_model_path)

        print(f"Epoch {epoch + 1} | lr={lr} | Train Loss: {epoch_train_losses[-1]:.4f} | "
              f"Val Loss: {val_loss:.4f} | Val Accuracy: {val_accuracy * 100:.2f}%")
````

````python
# Affichage des courbes d'évolution de la loss d'entraînement, loss de validation et accuracy de validation

plt.figure(figsize=(15, 5))

# Courbe de la loss d'entraînement
plt.subplot(1, 3, 1)
plt.plot(train_losses, label=f'lr={lr}')
plt.xlabel('Batch')
plt.ylabel('Training Loss')
plt.title(f'Training Loss per Batch (lr={lr})')
plt.legend()

# Courbe de la loss de validation
plt.subplot(1, 3, 2)
plt.plot(val_losses, label=f'lr={lr}')
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.title(f'Validation Loss per Epoch (lr={lr})')
plt.legend()

# Courbe de l'accuracy de validation
plt.subplot(1, 3, 3)
plt.plot(val_accuracies, label=f'lr={lr}')
plt.xlabel('Epoch')
plt.ylabel('Validation Accuracy')
plt.title(f'Validation Accuracy per Epoch (lr={lr})')
plt.legend()

plt.show()
````

**Q16**. Pourquoi est-il important de maintenir un ensemble de test séparé de l'ensemble de validation? Que se passerait-il si on n'avait que 2 ensembles (entraînement et test) plutôt que 3 (entraînement, validation, test)?

Parce qu'il faut assurer la bonne généralisation du modèle, il est important d'avoir à disposition un jeu de données de test et de validation. L'ensemble de validation est utile pendant la phase d'entraînement pour ajuster les hyperparamètres et faire le choix de la meilleure version du modèle, on teste donc différentes configurations sans affecter le résultat final. Une fois le modèle optimisé et totalement entraîné, l'ensemble de test sert pour une évaluation définitive sur la performance du modèle sur de nouvelles données.

**Q17**. Au sein de l'ensemble de test, identifiez
- Un exemple d'un vrai positif (une image cancéreuse correctement identifiée comme cancéreuse).
- Un exemple d'un faux positif (une image non-cancéreuse incorrectement identifiée comme cancéreuse).
- Un exemple d'un faux négatif (une image cancéreuse incorrectement identifiée comme non-cancéreuse).
- Un exemple d'un vrai négatif (une image non-cancéreuse correctement identifiée comme non-cancéreuse).

Affichez chaque image, son label, la prédiction de votre modèle, et l'index de l'exemple.

Afin qu'il sache comment sont retournées les prédictions, nous avons donné la fonction validate à chatGPT et nous lui avons posé la question telle quelle.

````python
# Initialiser les listes pour les différents types d'exemples
true_positives, false_positives, false_negatives, true_negatives = [], [], [], []

# Identifier les exemples en comparant prédictions et labels réels
for idx, batch in enumerate(test_loader):
    images, labels = batch["image"], batch["label"]
    for i in range(len(labels)):
        true_label = labels[i].item()
        pred_label = preds[idx * len(labels) + i]  # Accès à la prédiction correspondante

        if true_label == 1 and pred_label == 1:
            true_positives.append((images[i], true_label, pred_label, idx * len(labels) + i))
        elif true_label == 0 and pred_label == 1:
            false_positives.append((images[i], true_label, pred_label, idx * len(labels) + i))
        elif true_label == 1 and pred_label == 0:
            false_negatives.append((images[i], true_label, pred_label, idx * len(labels) + i))
        elif true_label == 0 and pred_label == 0:
            true_negatives.append((images[i], true_label, pred_label, idx * len(labels) + i))

# Fonction pour afficher les exemples
def display_example(example, title):
    image, true_label, pred_label, index = example
    image = image.cpu().squeeze().permute(1, 2, 0)  # Transformation pour l'affichage
    plt.imshow(image, cmap="gray")
    plt.title(f"{title}\nLabel: {true_label}, Prédiction: {pred_label}, Index: {index}")
    plt.axis("off")
    plt.show()

# Affichage de chaque catégorie d'exemple
if true_positives:
    display_example(true_positives[0], "Vrai Positif")
if false_positives:
    display_example(false_positives[0], "Faux Positif")
if false_negatives:
    display_example(false_negatives[0], "Faux Négatif")
if true_negatives:
    display_example(true_negatives[0], "Vrai Négatif")
````
